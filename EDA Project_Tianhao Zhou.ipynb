#  Hotel Booking EDA Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Total_Number_People']=df['adults']+df['children']+df['babies']
df['Not Adults']=df['children']+df['babies']
df['Total_Nights']=df['stays_in_weekend_nights']+df['stays_in_week_nights']
```

```python
fig, axes = plt.subplots(figsize=(12,6))

sns.countplot(x='arrival_date_month',data=df,palette='viridis')
```

```python
a = df['is_canceled'].groupby(df['arrival_date_month']).sum()
```

```python
fig, axes = plt.subplots(figsize=(12,6))
plt.bar(a.index,a,color='green')
plt.xlabel('Month')
plt.ylabel('Canceled')
```

```python
b=df[df['is_canceled']==0].groupby(df['arrival_date_month']).count()
```

```python
fig, axes = plt.subplots(figsize=(12,6))
plt.bar(b.index,b['is_canceled'],color='green')
plt.xlabel('Month')
plt.ylabel('Not Canceled')
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.countplot(x='arrival_date_month',data=df,hue='is_canceled',palette='viridis')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('df.corr()')
```

```python
is_repeated=df[df['is_repeated_guest']==1].groupby(df['country']).count()
df1=is_repeated.sort_values(by='hotel',ascending=False).head(10)
fig, axes = plt.subplots(figsize=(12,6))
plt.bar(df1.index,df1['hotel'],color='red')
plt.xlabel('Country')
plt.ylabel('Repeated Guest')
plt.title('Top 10 Country with Repeated Guest')
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.boxplot(x='arrival_date_month',y='lead_time',data=df,palette='rainbow')
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.boxplot(x='Not Adults',y='Total_Nights',data=df,palette='rainbow')
```

```python

sns.kdeplot(df['lead_time'][df['is_canceled'] == 0], shade= True)
sns.kdeplot(df['lead_time'][df['is_canceled'] == 1], color = "Red", shade=True)

plt.xlim([0, 800])
plt.xlabel("Lead Time")
plt.ylabel("Frequency")
plt.legend(['Not Canceled', 'Canceled'])

plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
market_size = df['market_segment'].value_counts().tolist()
labels = df['market_segment'].value_counts().index.tolist()

pie = plt.pie(market_size,autopct='%1.1f%%', startangle=90)
plt.legend(pie[0],labels,loc="best")
plt.title('Market Segment')

plt.show()
```

```python
sns.lineplot(x='meal', y='adr', data=df, hue='hotel')

plt.title("Meal Type and Hotel Rate")
plt.xlabel("Meal Type")
plt.ylabel("Rate")

plt.show()
```

```python
# Monthly lead_time:
df['lead_time_monthly'] = df['lead_time'] // 30
sns.lineplot(x='lead_time_monthly', y='adr', data=df, hue='hotel')

plt.title("Lead time vs ADR")
plt.legend(['resort', 'city'])
plt.xlabel("Lead Time (Month)")
plt.ylabel("Rate")

plt.show() 
```
